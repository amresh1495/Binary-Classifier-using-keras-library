<b>A basic binary classification model using Sequential() of the keras library. A generous implementation of Deep Learning consisting of only 2 hidden layers. 

```python
# importing required packages 

from keras.models import Sequential
from keras.layers import Dense
import numpy 
```

```python
# using seed() so that we get the same result everytime 

seed = 7
numpy.random.seed(seed)
```

```python
# importing the required data using panadas read_csv()

from pandas import read_csv
filename = 'C:\\Users\\amres\\Desktop\\Files\\BBCN.csv'
dataframe = read_csv(filename)
dataframe.head()
```

```python
# splitting the data into X and Y - input and output 

array = dataframe.values
X = array[:, 0:11]
Y = array[:, 11]
```

```python
# defining the model using Sequential() with 2 hidden layers, 11 inputs and 1 output 
# relu and signmoid are used as activation functions here initialised with uniform weights 

model = Sequential()
model.add(Dense(11, input_dim = 11, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# compliling the model defining the loss function, optimizer used and metric to be calculated 

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# fitting the model to the data 

model.fit(X, Y, epochs = 396, batch_size = 10)
```
